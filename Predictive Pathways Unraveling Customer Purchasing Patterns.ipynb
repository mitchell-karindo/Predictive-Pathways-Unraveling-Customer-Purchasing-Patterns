```python
#pemilihan x
X = df[['Education', 'Income', 'NumWebPurchases' ,'NumWebVisitsMonth','NumCatalogPurchases']]
print(X)
```

```python
#Melakukan endcoder pada education
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
X ['Education'] = le.fit_transform(X['Education'])
print(X)
```

```python
#penggunaan k means cluster
from sklearn.cluster import KMeans
```

```python
k = 4
```

```python
kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Education')
plt.ylabel('Income')
plt.legend()
plt.show()
```

```python
#penggunaan silhouette score
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
silhouette_scores = []

for k in range(2, 8):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)
    print(f"Silhouette Score untuk k={k}: {score:.3f}")
```

```python
# Plot Silhouette Score
plt.figure(figsize=(10, 6))
plt.plot(range(2, 8), silhouette_scores, marker='o')
plt.title('Silhouette Score untuk Menentukan Cluster')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
#dari hasil silhouette score ambil k-3
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
kmeans.fit(X)
```

```python
# menambah label cluster ke dataframe
df['Cluster'] = kmeans.labels_
```

```python
# Rata-rata karakteristik per cluster
cluster_summary = df.groupby('Cluster').agg({
    'Income': 'mean',
    'NumWebPurchases': 'mean',
    'NumCatalogPurchases': 'mean',
    'NumStorePurchases': 'mean',
    'MntWines': 'mean',
    'MntFruits': 'mean',
    'MntMeatProducts': 'mean',
    'MntFishProducts': 'mean',
    'MntSweetProducts': 'mean',
    'MntGoldProds': 'mean',
    'Cluster': 'count'
}).rename(columns={'Cluster': 'Count'})

print("\nRingkasan Cluster:")
print(cluster_summary)
```

```python
# Visualisasi cluster
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Income', y='NumWebPurchases', hue='Cluster', palette='viridis', s=100)
plt.title('Segmentasi pelanggan  dari  pendapatan dan pembelian')
plt.show()
```

```python
# Gabungan data demografis dengan cluster
cluster_profile = df.groupby('Cluster').agg({
    'Year_Birth': 'mean',
    'Income': 'mean',
    'Kidhome': 'mean',
    'Teenhome': 'mean',
    'NumWebPurchases': 'mean',
    'NumCatalogPurchases': 'mean',
    'NumStorePurchases': 'mean',
    'NumWebVisitsMonth': 'mean',
    'Response': 'mean',
    'MntWines': 'mean',
    'MntMeatProducts': 'mean'
})

print("\nProfil Cluster:")
print(cluster_profile)
```

# Rekomendasi strategi pemasaran:

1. Untuk Cluster 0 (Low Value):
   - Fokus pada program web store berikan rekomendasi yang sering mereka lirik atau favorit
   - Tawarkan produk yang sering mereka liat pada web store dengan harga lebih terjangkau

2. Untuk Cluster 1 (High Spenders):
   - Berikan penawaran premium dan produk high-end khususnya produk wine
   - Program diskon khusus untuk pembelian wine pada kelipatan tertentu
   - Personalisasi rekomendasi produk

3. Untuk Cluster 2 (Active Customers):
   - Program rewards untuk mempertahankan loyalitas
   - Rekomendasi produk berdasarkan riwayat pembelian
   - Tawarkan bundle produk harian/bulanan contohnya paket resep makanan

4. Untuk Cluster 3 (Inactive Customers):
   - website store arahkan untuk pembelian langsung agar customer mau beli offline
   - Penawaran khusus untuk pelanggan yang kembali
   - berikan diskon" khusus pelanggan yang kembali  di websitenya conth:claim diskon via website dan ambil langsung ke toko offline
